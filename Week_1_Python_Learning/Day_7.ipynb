````python
from random import choice


class Student:
    def __init__(self,name,rollno,age,marks):
        self.name = name
        self.rollno = rollno
        self.age = age
        self.marks = marks
    def calculate_average(self):
        return sum(self.marks)/len(self.marks)
    def total_marks(self):
        return sum(self.marks)
    def get_grade(self):
        avg=self.calculate_average()
        if avg >= 90: return "A"
        elif avg>=80 : return "B"
        elif avg >=70 :return "c"
        elif avg>=50 :return "D"
        else : return "Faile"
    def get_display(self):
        print(f"studen name = {self.name}")
        print(f"student ROll no ={ self.rollno}")
        print(f"student marks ={self.marks}")
        print(f"student  age ={self.age}")
        print(f"student avrage marks = {self.calculate_average()}")
        print(f"student total marks =  {self.total_marks()}/500")
        print(f"students grade = {self.get_grade()}.2f")
    def to_string(self):
        marks_str=",".join(map(str,self.marks))
        return f"{self.rollno},{self.name},{self.age},{marks_str}"
def loadStudents():
    students=[]
    try:
        with open("students.txt","r") as file:
            for line in file:
                data = line.strip().split(',')
                name=data[0]
                rollno=data[1]
                age=int(data[2])
                marks=[int(m)for m in data[3:]]
                students.append(Student(name,rollno,age,marks))
    except FileNotFoundError:
        pass
    return students

def save_students(students):
    with open("students.txt",'W') as file:
        for student in students:
            file.write(student.to_string()+ "\n")
def add_students(students):
    try:
        print("=============== Add students ==============")
        name=input("Enter Name")
        rollno=int(input("Eter Roll No"))
        age=int(input("Enter Age"))
        if age <=0:
            print("age must be positive number ")
            return
        
        marks=[]

        subjects= ["Math", "Science", "English", "Hindi", "Social"]
        for subject in subjects:
            mark=int(input(f"{subject} marks (0-100): "))
            if 0<= marks <=100:
                 marks.append(mark)
            else:
                print("marks must between 0-100")
                return 
        student=Student(name,rollno,age,marks)
        students.append(student)
        save_students(students)
        print("âœ… Student added successfully!")
    except ValueError:        
                print(" Invalid input!")
def view_all_students(students):
    if not students:
        print("not student found")
        return
    print("======== view all students info ========")
    for student in students:
        student.get_display()
def search_student(students):
    rol_no=input()
    for student in students:
        if student.rollno==rol_no:
            student.get_display()
            return
    print("âŒ Student not found!")
def top_student(students):
    if not students:
        print("not student found")
        # top 3 students
    sorted_students=sorted(students,key= lambda x:x.calculate_average(),reverse=True)
    for i ,sorted_students in enumerate(sorted_students[:3],start=1):
        print(f"Top {i} student:")
        sorted_students.get_display()

    # sorted_students=sorted(students,key= lambda x:x.calculate_average(),reverse=True)
    #     # top 3 student
    # for i in sorted_students[:3]:
    #     i.get_display()
def statistics(students):
    all_avg=[s.calculate_average() for s in students ]
    class_avg=sum(all_avg)/len(all_avg)
    passed= len([s for s in students if s.calculate_average()>=60])
    passed_percentage=(passed/len(students))*100

    highest= max(students,key=lambda x:x.calculate_average())
    lowest= min(students,key=lambda x:x.calculate_average())
    print("\n--- Class Statistics ---")
    print(f"Total Students: {len(students)}")
    print(f"Class Average: {class_avg:.2f}")
    print(f"Pass Percentage: {passed_percentage:.2f}%")
    print(f"\nHighest: {highest.name} - {highest.calculate_average():.2f}")
    print(f"Lowest: {lowest.name} - {lowest.calculate_average():.2f}")
def main():
    """Main program"""
    print("\n" + "="*50)
    print(" "*10 + "STUDENT MANAGEMENT SYSTEM")
    print("="*50)
    students=loadStudents()
    print(f"\nâœ… Loaded {len(students)} students from file.")
    
    while True:
        print("\n--- MENU ---")
        print("1. Add Student")
        print("2. View All Students")
        print("3. Search Student")
        print("4. Top 3 Students")
        print("5. Class Statistics")
        print("6. Exit")
        try:
            choice = int(input("\nEnter choice: "))
            if choice==1:
                add_students(students)
            elif choice ==2:
                view_all_students(students)
            elif choice ==3:
                search_student(students)

            elif choice == 4:
                top_student(students)
            elif choice ==5:
                statistics(students)
            elif choice == 6:
                print("\nâœ… Thank you! Goodbye!")
                break
            else:
                print("âŒ Invalid choice!")
        except ValueError:
            print("âŒ Please enter a number!")
if __name__ == "__main__":
    main()
````

---

### What is SQL?

**SQL = Structured Query Language**
- Language for managing databases
- Store, retrieve, update data
- Used EVERYWHERE (websites, apps, analytics)

---

### Why SQL for Backend/Data Engineering?

**Backend Developer:**
- Store user data (accounts, posts, comments)
- Authenticate users
- Manage application data

**Data Engineer:**
- Extract data from databases
- Transform and clean data
- Load into data warehouses

---

### Basic SQL Concepts

#### 1. Database
Container for tables. Like a folder.

#### 2. Table
Stores data in rows and columns. Like Excel sheet.

**Example: students table**
```
| id | name   | age | marks |
|----|--------|--------|-------|
| 1  | Raj    | 22  | 85    |
| 2  | Priya  | 21  | 92    |
| 3  | Amit   | 23  | 78    |
```

#### 3. Row (Record)
Single entry. One student.

#### 4. Column (Field)
Attribute. Like name, age.

---

### Basic SQL Commands

#### SELECT - Read data
```sql
-- Get all students
SELECT * FROM students;

-- Get specific columns
SELECT name, age FROM students;

-- Get with condition
SELECT * FROM students WHERE age > 21;

-- Get with sorting
SELECT * FROM students ORDER BY marks DESC;
```

#### INSERT - Add data
```sql
INSERT INTO students (name, age, marks) 
VALUES ('Neha', 20, 88);
```

#### UPDATE - Modify data
```sql
UPDATE students 
SET marks = 90 
WHERE name = 'Raj';
```

#### DELETE - Remove data
```sql
DELETE FROM students 
WHERE age < 18;
```
