### What is a Variable?

Think of a variable as a **labeled box** where you store data.

**Real life:**
- Box labeled "Phone" ‚Üí Contains your phone number
- Box labeled "Age" ‚Üí Contains your age

**In Python:**
```python
phone = "9876543210"    # Box labeled 'phone'
age = 22                # Box labeled 'age'
```

---

### Rules for Variable Names:

**‚úÖ ALLOWED:**
```python
name = "Raj"              # lowercase
student_age = 22          # underscore OK
marks_2024 = 85           # numbers OK (not at start)
isStudent = True          # camelCase OK
```

**‚ùå NOT ALLOWED:**
```python
2name = "Raj"             # Can't start with number
student-age = 22          # No hyphens
for = 10                  # Can't use Python keywords
my name = "Raj"           # No spaces
```

**‚ú® BEST PRACTICE:**
```python
# Use meaningful names
student_name = "Raj"      # Good - clear what it is
x = "Raj"                 # Bad - not clear

total_marks = 450         # Good
tm = 450                  # Bad
```

### Data Types (Types of Data)

#### 1. String (str) - Text with quotes
```python
name = "Raj Kumar"              # Double quotes
city = 'Ahmedabad'              # Single quotes (same thing)
message = "Hello!"

# Check type
print(type(name))               # <class 'str'>
```

#### 2. Integer (int) - Whole numbers, NO quotes
```python
age = 22
marks = 85
year = 2026

print(type(age))                # <class 'int'>
```

#### 3. Float - Decimal numbers
```python
price = 99.99
temperature = 36.6
pi = 3.14159

print(type(price))              # <class 'float'>
```

#### 4. Boolean (bool) - True/False ONLY (Capital T and F)
```python
is_student = True
has_job = False

print(type(is_student))         # <class 'bool'>
```

---

### Type Conversion (Convert between types)

```python
# String to Integer
age_text = "22"
age_number = int(age_text)      # "22" becomes 22
print(age_number + 1)           # 23 (math works now!)

# Integer to String
age = 22
age_text = str(age)             # 22 becomes "22"
print("Age: " + age_text)       # Can join with string now

# String to Float
price = "99.99"
price_num = float(price)        # "99.99" becomes 99.99

# Integer to Float
num = 10
num_decimal = float(num)        # 10 becomes 10.0
```

**‚ö†Ô∏è WHY YOU NEED THIS:**
```python
# This will ERROR:
age = "22"                      # This is text!
next_year = age + 1             # ERROR! Can't add text + number

# FIX:
age = int("22")                 # Convert to number first
next_year = age + 1             # Now works! 23
```

### Output - print() function

Shows information to the user.

```python
# Print text
print("Hello World")

# Print variable
name = "Raj"
print(name)                     # Raj

# Print multiple things (auto adds space)
age = 22
print("Name:", name, "Age:", age)    # Name: Raj Age: 22

# Join with + (ONLY strings!)
print("Hello " + name)          # Hello Raj

# Can't mix types with +
# print("Age: " + age)          # ERROR!
print("Age: " + str(age))       # Age: 22 (converted to string first)
```

**üéØ F-STRINGS (Modern, Easy, BEST WAY!):**
```python
name = "Raj"
age = 22

# Put 'f' before quote, use {} for variables
print(f"My name is {name} and I am {age} years old")
# Output: My name is Raj and I am 22 years old

# Can do math inside {}
print(f"Next year I'll be {age + 1}")
# Output: Next year I'll be 23

# Can use methods inside {}
print(f"Uppercase name: {name.upper()}")
# Output: Uppercase name: RAJ
```

### Arithmetic Operators (Math)

```python
a = 10
b = 3

print(a + b)    # Addition: 13
print(a - b)    # Subtraction: 7
print(a * b)    # Multiplication: 30
print(a / b)    # Division: 3.333... (always gives float)
print(a // b)   # Floor Division: 3 (integer part only)
print(a % b)    # Modulo (remainder): 1 (10 √∑ 3 = 3 remainder 1)
print(a ** b)   # Power: 1000 (10¬≥ = 10√ó10√ó10)
```

**üéØ MODULO (%) IS SUPER USEFUL:**
```python
# Check if number is even or odd
num = 10
if num % 2 == 0:    # If remainder is 0, it's even
    print("Even")
else:
    print("Odd")
```

---

### Comparison Operators (Returns True/False)

```python
a = 10
b = 20

print(a == b)   # Equal to: False (use == not =)
print(a != b)   # Not equal: True
print(a > b)    # Greater than: False
print(a < b)    # Less than: True
print(a >= 10)  # Greater or equal: True
print(a <= 10)  # Less or equal: True
```

**‚ö†Ô∏è COMMON MISTAKE:**
```python
# WRONG:
if age = 18:        # This assigns value! ERROR!

# CORRECT:
if age == 18:       # This compares! Use ==
```

---

### Assignment Operators (Shortcuts)

```python
x = 10

x = x + 5       # Same as: x += 5
x += 5          # Add 5 to x (x becomes 15)

x -= 3          # Subtract 3 from x
x *= 2          # Multiply x by 2
```

## üìñ PART 4: Strings (Read: 40 min) - VERY IMPORTANT!

---

### Creating Strings

```python
# Single or double quotes (both same)
name = "Raj"
city = 'Ahmedabad'

# Multi-line strings (use triple quotes)
message = """Hello,
This is a message
on multiple lines."""

# Empty string
empty = ""
```

---

### String Concatenation (Joining)

```python
first = "Raj"
last = "Kumar"

# Join with +
full = first + " " + last
print(full)         # Raj Kumar

# Repeat with *
print("Ha" * 3)     # HaHaHa
print("-" * 20)     # --------------------
```

---

### String Length

```python
text = "Hello World"
length = len(text)      # 11 (includes space)
print(length)

# Empty string
empty = ""
print(len(empty))       # 0
```

---

### String Indexing (Getting Single Character)

**‚ö†Ô∏è IMPORTANT: Python counts from 0!**

```python
text = "Python"
#       012345    <- forward index
#      -6-5-4-3-2-1  <- backward index

# Forward indexing
print(text[0])          # P (first)
print(text[1])          # y
print(text[5])          # n (last)

# Backward indexing (from end)
print(text[-1])         # n (last)
print(text[-2])         # o (second last)
print(text[-6])         # P (first)
```

---

### String Slicing (Getting Part of String)

**Syntax:** `string[start:end]`  
- start is INCLUDED
- end is EXCLUDED

```python
text = "Hello World"
#       01234567891011

print(text[0:5])        # Hello (0,1,2,3,4 - stops before 5)
print(text[6:11])       # World

# Shortcuts
print(text[:5])         # Hello (from start to 5)
print(text[6:])         # World (from 6 to end)
print(text[:])          # Hello World (entire string)

# With step (every nth character)
print(text[::2])        # HloWrd (every 2nd)
print(text[::-1])       # dlroW olleH (REVERSE!)
```

**üéØ PRACTICAL EXAMPLE:**
```python
email = "raj.kumar@gmail.com"

# Get username (before @)
at_pos = email.find("@")        # Find @ position
username = email[:at_pos]       # Slice up to @
print(username)                 # raj.kumar

# Get domain (after @)
domain = email[at_pos+1:]       # Slice after @
print(domain)                   # gmail.com
```

### String Methods (Built-in Functions)

#### 1. Case Conversion
```python
text = "Hello World"

print(text.upper())     # HELLO WORLD
print(text.lower())     # hello world
print(text.title())     # Hello World (first letter caps)
print(text.capitalize()) # Hello world (only first caps)

# Original doesn't change!
print(text)             # Hello World (still same)
```

#### 2. Strip (Remove spaces/characters)
```python
text = "  Hello  "

print(text.strip())     # "Hello" (both sides)
print(text.lstrip())    # "Hello  " (left only)
print(text.rstrip())    # "  Hello" (right only)
```

#### 3. Replace
```python
text = "I love Python"

new = text.replace("Python", "Coding")
print(new)              # I love Coding

# Replaces ALL occurrences
text = "hello hello hello"
new = text.replace("hello", "hi")
print(new)              # hi hi hi
```

#### 4. Split (Break into list)
```python
# Split by comma
text = "apple,banana,cherry"
fruits = text.split(",")
print(fruits)           # ['apple', 'banana', 'cherry']

# Split by space (default)
sentence = "Hello World Python"
words = sentence.split()
print(words)            # ['Hello', 'World', 'Python']
```

#### 5. Find / Count
```python
text = "Hello World"

# Find position
pos = text.find("World")        # 6
print(pos)

# If not found
pos = text.find("Python")       # -1 (not found)
print(pos)

# Count occurrences
text = "banana"
count = text.count("a")         # 3
print(count)
```

#### 6. Check contents
```python
text = "Hello World"

# Check if substring exists
print("Hello" in text)          # True
print("Python" in text)         # False

# Starts/ends with
print(text.startswith("Hello")) # True
print(text.endswith("World"))   # True

# Check character types
print("123".isdigit())          # True (all numbers)
print("abc".isalpha())          # True (all letters)
print("abc123".isalnum())       # True (letters + numbers)
```
